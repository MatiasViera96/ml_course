```python
# How about we try another type of plot? This time let's look at a bar plot. 
# First we'll make some data.

# Create a Python dictionary of 3 of your favourite foods with 
# The keys of the dictionary should be the food name and the values their price
food_dictionary = {"Pizza":100, "Hamburguesa": 150, "Milanesa": 160};
food_dictionary
```

```python
# Create a bar graph where the x-axis is the keys of the dictionary
# and the y-axis is the values of the dictionary
fig_food, ax_food = plt.subplots();
ax_food.bar(food_dictionary.keys(), food_dictionary.values())

# Add a title, xlabel and ylabel to the plot
ax_food.set(title = "Food Dictionary Bar Plot",
            xlabel = "Food",
            ylabel = "Price ($)");
plt.show();
```

```python
# Make the same plot as above, except this time make the bars go horizontal
fig_fod_hor,ax_food_hor = plt.subplots()
ax_food_hor.barh(list(food_dictionary.keys()), list(food_dictionary.values()))
ax_food_hor.set(title = "Food Dictionary horizontal Bar Plot",
            ylabel = "Food",
            xlabel = "Price ($)");
plt.show();
```

```python
# Import pandas as pd
import pandas as pd;
```

```python
# Import the '../data/car-sales.csv' into a DataFame called car_sales and view
car_sales = pd.read_csv('car-sales.csv');
car_sales
```

```python
# Try to plot the 'Price' column using the plot() function
fig_car_sales, ax_car_sales = plt.subplots();
car_sales["Price"].plot();
```

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
car_sales['Price'] = car_sales['Price'].str.replace("$","").str.replace(",","").str.replace(".00","");
car_sales['Price'] = car_sales['Price'].astype('int32');
car_sales['Price']
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
car_sales["Total Sales"] = car_sales["Price"].cumsum(axis=None, skipna = True);

# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
car_sales["Sale Date"] = pd.date_range(start = "2024-05-01", end = "2024-05-10", freq = "D");

# View the car_sales DataFrame
car_sales
```

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
plt.plot(car_sales["Sale Date"],car_sales["Total Sales"]);
```

```python
# Convert the 'Price' column to the integers
# Already did it

# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function
car_sales.plot(x = "Odometer (KM)", y = "Price", kind = "scatter");
```

```python
# Import "../data/heart-disease.csv" and save it to the variable "heart_disease"
heart_disease = pd.read_csv("heart-disease.csv");
```

```python
# View the first 10 rows of the heart_disease DataFrame
heart_disease[:10]
```

```python
# Create a histogram of the "age" column with 50 bins
heart_disease["age"].plot(kind = "hist", bins = 50)
```

```python
# Call plot.hist() on the heart_disease DataFrame and toggle the
# "subplots" parameter to True
#heart_disease.plot.hist(subplots = True)

# That plot looks pretty squished. Let's change the figsize.

# Changing figsize to (10,30)
heart_disease.plot.hist(subplots = True, figsize = (10,30))
```

```python
# Replicate the above plot in whichever way you see fit

# Note: The method below is only one way of doing it, yours might be
# slightly different

# Create DataFrame with patients over 50 years old
patients_over_50 = heart_disease[heart_disease["age"] >= 50];

# Create the plot
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize =(10,5))

# Plot the data
scatter = ax.scatter(patients_over_50["age"],
           patients_over_50["chol"],
           c = patients_over_50["target"])

# Customize the plot
ax.set_xlabel("Age")
ax.set_ylabel("Cholesterol")
ax.set_title("HeartDisease and Cholesterol Levels")
ax.legend(*scatter.legend_elements(), title = "Target")

# Add a meanline
ax.axhline(patients_over_50["chol"].mean(),
           linestyle = "--")
```

```python
# this is not looking good. Lets try to change the colors a little
# Replot the same figure as above except change the "cmap" parameter
# of scatter() to "winter"
# Also change the "color" parameter of axhline() to "red"

# Create the plot
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize =(10,5))

# Plot the data
scatter = ax.scatter(patients_over_50["age"],
           patients_over_50["chol"],
           c = patients_over_50["target"],
           cmap = "winter")

# Customize the plot
ax.set_xlabel("Age")
ax.set_ylabel("Cholesterol")
ax.set_title("Heart Disease and Cholesterol Levels")
ax.legend(*scatter.legend_elements(), title = "Target")

# Add a meanline
ax.axhline(patients_over_50["chol"].mean(),
           linestyle = "--",
           color = "red")
```
