```python
# Import the car sales DataFrame with missing data ("../data/car-sales-missing-data.csv")

car_sales_missdata_df = pd.read_csv("car-sales-missing-data.csv")

# Check out the new DataFrame
car_sales_missdata_df
```

```python
# Fill the Odometer column missing values with the mean of the column inplace
car_sales_missdata_df.Odometer.fillna(car_sales_missdata_df.Odometer.mean(), inplace = True)
```

```python
# View the car sales missing DataFrame and verify the changes
car_sales_missdata_df
```

```python
# Changing price column in car_sales_missdata_df to int64
#car_sales_missdata_df.Price = car_sales_missdata_df.Price.str.replace("$","",).str.replace(",","")
#car_sales_missdata_df.Price.fillna("0",inplace=True)
#car_sales_missdata_df.Price = car_sales_missdata_df.Price.astype("int64")
car_sales_missdata_df
```

```python
# Filling nan (or 0 values) with mean
car_sales_missdata_df.Price.replace(0,pd.NA,inplace= True)
car_sales_missdata_df.Price.fillna(car_sales_missdata_df.Price.mean(),inplace = True)
```

```python
# Checking the changes where applied correctly
car_sales_missdata_df
```

```python
# changing doors nan values to mean
car_sales_missdata_df.Doors.fillna(car_sales_missdata_df.Doors.mean(), inplace = True)
```

```python
car_sales_missdata_df
```

```python
# Finding the least repeated colours
car_sales_missdata_df.Colour.value_counts()
```

```python
#Filling nan colours with red
car_sales_missdata_df.Colour.fillna("Red", inplace = True)
car_sales_missdata_df
```

```python
# Finding the least repeated car maker
car_sales_missdata_df.Make.value_counts()
```

```python
# Filling nan makers with Nissan
car_sales_missdata_df.Make.fillna("Nissan", inplace = True)
car_sales_missdata_df
```

```python
# Create a "Seats" column where every row has a value of 5
car_sales_missdata_df["Seats"] = 5
car_sales_missdata_df
```

```python
# Importing numpy
import numpy as np
# Create a column called "Engine Size" with random values between 1.3 and 4.5
car_sales_missdata_df["Engine Size"] = np.random.uniform(low = 1.3, high = 4.5, size = len (car_sales_missdata_df)).round(1)
car_sales_missdata_df
```

```python
# Create a column which represents the price of a car per kilometer
car_sales_missdata_df["Price per KM"] = car_sales_missdata_df.Price / car_sales_missdata_df.Odometer
car_sales_missdata_df
```

```python
# Remove the last column you added using .drop()
car_sales_missdata_df.drop(columns=["Price per KM"], inplace = True)
car_sales_missdata_df
```

```python
# Shuffle the DataFrame using sample() with the frac parameter set to 1
# Save the the shuffled DataFrame to a new variable
car_sales_missdate_shuffled_df = car_sales_missdata_df.sample(frac=1)
car_sales_missdate_shuffled_df
```

```python
# Reset the indexes of the shuffled DataFrame
car_sales_missdate_shuffled_df.reset_index(drop = True, inplace = True)
car_sales_missdate_shuffled_df
```

```python
# Change the title of the Odometer (KM) to represent miles instead of kilometers
car_sales_missdata_df.rename(columns = {"Odometer": "Odometer (mi)"}, inplace = True)

# Change Odometer column values from Km to mi
car_sales_missdata_df["Odometer (mi)"] = car_sales_missdata_df["Odometer (mi)"] * 0.6214
# Round values to 1 digit afther the ","
car_sales_missdata_df["Odometer (mi)"] = np.round(car_sales_missdata_df["Odometer (mi)"],1)
#Show data changes
car_sales_missdata_df
```
